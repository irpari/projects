```python
import pandas as pd
from sklearn import tree
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import normalize
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, plot_confusion_matrix, roc_auc_score
import matplotlib.pyplot as plt
```

# Data classification w decision tree

```python
df = pd.read_csv("bank_dataset_cleaned.csv", index_col=0)
```

```python
# Convert categorical variables to dummies
df_w_dummies = pd.get_dummies(data=df, columns = ["job", "marital", "poutcome", "month", "contact"], \
                              prefix = ["job", "marital", "poutcome", "month", "contact"])
df_w_dummies.head()
```

```python
x_data = df_w_dummies.copy().drop(columns=["deposit"])
y_data = df_w_dummies["deposit"].copy()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, 
                                                    test_size=0.4, 
                                                    random_state=777, 
                                                    stratify=y_data)
# Further splitting for validation
x_test, x_val, y_test, y_val = train_test_split(x_test, 
                                                y_test, 
                                                test_size=0.5, 
                                                random_state=777, 
                                                stratify=y_test)
```

```python
sample_leaf = [300, 150, 125, 100, 75, 50, 25, 15]
for min_leaf in sample_leaf:
    clf_tmp = tree.DecisionTreeClassifier(criterion="gini", min_samples_leaf=min_leaf)
    model = clf_tmp.fit(x_train, y_train)
    y_pred = model.predict(x_val)
    print("Min leaf: {:3d} \t Accuracy: {:.3f} \t Recall: {:.3f}".format(min_leaf, 
                                                                         accuracy_score(y_val,y_pred), 
                                                                         recall_score(y_val, y_pred)))
```

```python
clf_tree = tree.DecisionTreeClassifier(criterion="gini", min_samples_leaf=75, random_state=777)
model = clf_tree.fit(x_train, y_train)
y_pred = model.predict(x_test)
print("Best model \t Accuracy: {:.3f} \t Recall: {:.3f} \t\t ROC: {:.3f}"
      .format(accuracy_score(y_test,y_pred), 
              recall_score(y_test, y_pred),
              roc_auc_score(y_test, y_pred)))
plot_confusion_matrix(clf_tree, x_test, y_test);
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (20, 20), dpi=300)
tree.plot_tree(clf_tree, feature_names=x_data.columns);
fig.suptitle("Decision tree", fontweight="bold", fontsize=24)
fig.savefig("tree.png")
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (20, 20), dpi=300)
tree.plot_tree(clf_tree, max_depth=2, feature_names=x_data.columns);
fig.suptitle("Decision tree, upper branches", fontweight="bold", fontsize=24)
fig.savefig("tree_cropped.png")
```

```python
# Extract feature importance
importance = pd.DataFrame({"feature": x_train.columns, "importance" : clf_tree.feature_importances_})
importance.sort_values("importance", ascending=False, inplace = True)
print(importance)
```

# Data classification w nn classifier

```python
# Dropping: job, education, default, loan,, marital
df_cut = df.drop(columns=["job", "default", "loan"])
#df_cut = df.copy()
```

```python
# Encoding categories to numbers
ord_enc = OrdinalEncoder()
df_net = df_cut.copy()
df_net["contact"] = ord_enc.fit_transform(df_net[["contact"]])
df_net["poutcome"] = ord_enc.fit_transform(df_net[["poutcome"]])
#df_net["job"] = ord_enc.fit_transform(df_net[["job"]])
# df_net["education"] = ord_enc.fit_transform(df_net[["education"]])
# df_net["marital"] = ord_enc.fit_transform(df_net[["marital"]])
df_net = pd.get_dummies(data=df_net, columns = ["marital", "education"], \
                              prefix = ["marital", "education"])
```

```python
clf_nn = MLPClassifier(hidden_layer_sizes=(32, 32, 16, 16), 
                       random_state=777, 
                       max_iter=int(1e8), 
                       early_stopping=True, 
                       n_iter_no_change=1e3)
```

```python
x_data_net = df_net.drop(columns=["deposit"]).copy()
x_data_net = normalize(x_data_net, norm='max', axis=0, copy=True, return_norm=False)
y_data_net = df_net["deposit"].copy()
```

```python
x_train_net, x_test_net, y_train_net, y_test_net = train_test_split(x_data_net, 
                                                                    y_data_net, 
                                                                    test_size=0.2, 
                                                                    random_state=777, 
                                                                    stratify=y_data_net)
```

```python
clf_nn.fit(x_train_net, y_train_net)
```

```python
y_pred_net = clf_nn.predict(x_test_net)
```

```python
print("Best model \t Accuracy: {:.3f} \t Recall: {:.3f} \t\t ROC: {:.3f}"
      .format(accuracy_score(y_test_net,y_pred_net), 
              recall_score(y_test_net, y_pred_net),
              roc_auc_score(y_test_net, y_pred_net)))
plot_confusion_matrix(clf_nn, x_test_net, y_test_net);
```
